```python
import pandas as pd
import numpy as np
embeddings = pd.read_csv("/content/drive/MyDrive/emotion_flow_thesis/nrclexicon/embeddings.csv")
display(embeddings)
embeddings =  np.array(embeddings)
embeddings
```

```python
## conver array to C-continuous for Faiss ##
embeddings = embeddings.copy(order='C')
display(embeddings.flags)

## convert numpy array to float32 for Faiss ##
embeddings.dtype
embeddings = np.float32(embeddings)
embeddings.dtype

```

```python
d = embeddings.shape[1]
index = faiss.IndexFlatL2(d)  
index.add(embeddings)
```
