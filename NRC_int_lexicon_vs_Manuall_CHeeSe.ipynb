###### Create highest emotion vectors for each article based on its emotion flow produced from NRC intensity lexicon 

```python
emotion_ids_dict = {'nan':-1,'anger': 0, 'anticipation': 1, 'disgust': 2, 'fear': 3, 'joy': 4, 'sadness': 5,
       'surprise': 6, 'trust': 7}

emotion_ids_dict = {'nan':-1,'Anger': 0, 'Anticipation': 1, 'Disgust': 2, 'Fear': 3, 'Joy': 4, 'Sadness': 5,
       'Surprise': 6, 'Trust': 7}

def get_emotions_per_paragraph(article_ef_df):
  num_paragraphs = len(article_ef_df.columns)
  emotion_id_list = []
  for i in range(num_paragraphs):
    paragraph_vector_df = article_ef_df[[str(i)]]
    hiest_paragraph_emotion = str(paragraph_vector_df.idxmax().tolist()[0])
    avg = paragraph_vector_df.mean()
    if avg[0] == 0.0:
      hiest_paragraph_emotion = 'nan'
  
    emotion_id_list.append(emotion_ids_dict[hiest_paragraph_emotion])
  return emotion_id_list
```
